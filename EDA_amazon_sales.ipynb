**<h5><u>Preparing data**

<h4>Changing datatypes of a few columns:

```python
df['Order Date'] = pd.to_datetime(df['Order Date'], infer_datetime_format=True, errors='coerce') \
                    .copy()
df['Ship Date'] = pd.to_datetime(df['Ship Date'], infer_datetime_format=True, errors='coerce') \
                    .copy()
```

```python
df.columns
```

<h4>Dropping / Adding columns to dataframe for data analysis:

```python
df = df.drop(['Order ID'], axis=1).copy()
df
```

```python
del_time = df['Ship Date'] - df['Order Date']
df['Delivery Time (Days)'] = del_time.dt.days.copy()
df
```

**<h2 style = "text-align: center;"><u>Bivariate Analysis**

<h5>Checking correlations between few parameters with a correlation matrix:
<li> +1 -> Perfect Positive Relationship
<li> 0  -> No Relationship
<li> -1 -> Perfect Negative Relationship

```python
df_corr = df.select_dtypes(include=['int64', 'float64']) \
    .corr()
df_corr
```

<h5>Constructing a correlational heatmap for better visualization of strength between variables:

```python
sns.heatmap(data=df_corr, annot=True, linewidths=.5, cmap = 'coolwarm', fmt=".3f") \
.set_title('Correlation Heatmap of some parameters')
plt.xticks(rotation=45, ha='right')
plt.show()
```

<h5>Plotting scatterplots to verify all these correlations:

```python
sns.pairplot(data=df, hue='Sales Channel')
plt.show()
```

<h5>Plotting a few important relations (regression plots with 95% confidence intervals):

**<h4>1. Units Sold vs. Total Revenue**

```python
plt.figure(figsize=(10,5))

sns.lmplot(data=df, 
      x='Units Sold', 
      y='Total Revenue',
      hue='Sales Channel',
      robust=True)

plt.show()
```

<p style = "text-align: center;">Units Sold and Total Revenue has a <u>moderate correlation of ~0.45</u>, it contains many <u>outliers</u> suggesting that it may be <u>due to promotions or bulk purchases.</u>

**<h4>1. Units Sold vs. Total Profit**

```python
plt.figure(figsize=(10,50))

sns.lmplot(data=df, 
      x='Units Sold', 
      y='Total Profit',
      hue='Sales Channel',
      robust=True)

plt.show()
```

<p style = "text-align: center;">Units Sold and Total Profit has a <u>moderate correlation of ~0.57</u>, a considerable amount of <u>outliers below the regression line</u> suggest it could be <u>due to heavy offers and discounts</u> on these products which result in <u>lower profit margins.</u> <u>Outliers above the line</u> suggest that these <u>might be very niche-specific/luxury products.</u>

**<h4>1. Total Cost vs. Total Profit**

```python
plt.figure(figsize=(10,50))

sns.lmplot(data=df, 
      x='Total Cost', 
      y='Total Profit',
      hue='Sales Channel',
      robust=True)

plt.show()
```

<p style = "text-align: center;">Total Cost and Total Profit has a <u>strong correlation of ~0.8</u>, where specifically <u>profit from online purchases seems to be doing slightly better than offline purchases.</u> Many <u>outliers above the regression lines</u> suggests that these products are <u>highly profitable and require low cost to produce.</u>

**<h2 style="text-align: center;"><u>A FEW STATISTICAL QUESTIONS:**

<h4>1. Which region has made the largest profits?

```python
region_profit = df.groupby('Region')['Total Profit'].sum()

plt.figure(figsize=(10,5))

region_profit.plot(kind='barh', 
      xlabel='Total Profit', 
      ylabel='Region', 
      title='Profits made in each region between ' + str(df['Order Date'].min().year) + ' and ' + str(df['Order Date'].max().year),
      color='maroon')
plt.show()
```

**<p style = "text-align: center;">As we can see, the largest profits are made in <u>Sub-Saharan Africa</u> followed by <u>Europe</u> with a good majority too**

<h4>2. Is there any relation between the unit pricing and unit cost of these products? Does Sales Channel play a major role in this comparison?

```python
print(df[['Unit Price', 'Unit Cost']].corr())

plt.figure(figsize=(10,5))

sns.lmplot(data=df,
           x='Unit Price',
           y='Unit Cost',
           hue='Sales Channel')

plt.show()
```

**<p style = "text-align: center;">Seems that they have a <u>strong positive linear relationship</u> with a <u>correlation of ~0.99</u>, which suggests that the <u>profit margin is relatively stable</u> across different products. Also, the sales channel doesn't seem to make any difference for this profit margin.**


<h4>3. Which types of products have the largest revenue?

```python
item_revenue = df.groupby('Item Type')['Total Revenue'].sum()

plt.figure(figsize=(10,5))

item_revenue.plot(kind='bar', 
      xlabel='Item Type', 
      ylabel='Total Revenue', 
      title='Revenue made by each product between ' + str(df['Order Date'].min().year) + ' and ' + str(df['Order Date'].max().year),
      color='green')

plt.xticks(rotation=45, ha='right')
plt.show()
```

**<p style = "text-align: center;"><u>Cosmetics, Household items and Office Supplies</u> make up most of the revenue of these sales.**

<h4>4. Identify the relationship between order priority and delivery time of these products.

```python
df['Order Priority'].unique()
```

```python
plt.figure(figsize=(12,5))

m = df.groupby('Order Priority')['Delivery Time (Days)']

sns.kdeplot(data=m.get_group('H'), fill=True, color='blue', label='H')
sns.kdeplot(data=m.get_group('L'), fill=True, color='yellow', label='L')
sns.kdeplot(data=m.get_group('M'), fill=True, color='green', label='M')
sns.kdeplot(data=m.get_group('C'), fill=True, color='red', label='C') \
    .set_title('Delivery Time distribution based on Order Priority between '+ str(df['Order Date'].min().year) + ' and ' + str(df['Order Date'].max().year))

plt.legend(title='Order Priority')
plt.show()
```

**<p style = "text-align: center;">From this density plot, we can understand that:
    <li style="text-align: center;">Orders with <u>priority 'H'</u> take mostly around <u>20 days</u> to deliver
    <li style="text-align: center;">Orders with <u>priority 'L' and 'M'</u> have a bimodal distribution which says that majority of orders take around <u>5 days</u> as well as <u>40 days</u> to deliver
    <li style="text-align: center;">Orders with <u>priority 'C'</u> take mostly around <u>30 days</u> to deliver</p>**

<h4>5. Identify which Sales Channel is more profitable.

```python
plt.figure(figsize=(10,6))

profit_channel = df.groupby('Sales Channel')['Total Profit'].sum()

plt.pie(profit_channel, 
        labels=profit_channel.index, 
        autopct='%0.2f%%', 
        startangle=20, 
        explode=[0.06,0], 
        colors=['r','b'],
        shadow=True,
        radius=1.15,
        textprops={'fontsize':12},
        wedgeprops={'linewidth':5})

plt.title('Total Profit earned based on Sales Channel')
plt.show()
```

**<p style = "text-align: center;">Products sold through <u>offline channel</u> makes up majority <u>(56.42%)</u> of the sales profit.**
